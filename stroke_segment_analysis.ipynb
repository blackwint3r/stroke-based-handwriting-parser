```python
import struct
def hex2float(s):
    return struct.unpack('<f',bytes.fromhex(s))[0]
```

```python
cu.execute("select * from annotation")
ana_table = cu.fetchall()
print (ana_table)
```

```python
struct.unpack('14f',ana_table[0][-8])
```

```python
blobOfStroke = []
blobMat = []

for i in range(0, len(ana_table)):
    j = ana_table[i][-9]
    j2 = str(7*j)
    
    blobOfStroke.append(list(struct.unpack(j2+'f',ana_table[i][-8])))
    blobMat.append([])
    for j in range(0,j):
        blobMat[i].append(blobOfStroke[i][j*7:(j+1)*7])
print(blobOfStroke)
print(blobMat)
```
